```python
url = 'https://api.nomics.com/v1/volume/history'

start = '2019-07-06 00:00:00'
end = '2019-07-08 00:00:00'
start_unix = ts_to_unix(start)
end_unix = ts_to_unix(end)
start = unix_to_ts(start_unix)
end = unix_to_ts(end_unix)
start_rfc3339 = start.strftime('%Y-%m-%dT') + '00:00:00Z'
end_rfc3339 = end.strftime('%Y-%m-%dT') + '00:00:00Z'

parameters = {
    'key': 'e729d9cb5041a06c2a839122e0994bfe',
    'start': start_rfc3339,
    'end': end_rfc3339,
}
response = requests.get(url=url, params=parameters)
if response.status_code != 200:
    print(f"Response Status: {response.status_code}")

hist_vol = pd.DataFrame(response.json())
hist_vol.rename({'timestamp': 'Date', 'volume': 'Market Volume'},
                axis=1, inplace=True)
hist_vol['Date'] = hist_vol['Date'].apply(remove_tz)
hist_vol = hist_vol.set_index('Date')
hist_vol.head()
#hist_vol.to_csv(file_name)
```

```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d %H:%M:%S")
    print(d1)
    d2 = datetime.strptime(d2,  "%Y-%m-%d %H:%M:%S")
    return abs((d2 - d1).days + 1)
start = '2014-11-28 00:00:00'
end = '2021-05-10 00:00:00'
print(days_between(start,end))
```
